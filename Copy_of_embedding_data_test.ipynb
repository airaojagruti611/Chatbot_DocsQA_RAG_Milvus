```python
#test for Fingerhut dataset

from langchain_community.document_loaders import RecursiveUrlLoader

DOCS_PAGE="https://docs.nvidia.com/cuda/"

loader = RecursiveUrlLoader(DOCS_PAGE)
docs = loader.load()

num_documents = len(docs)
print(f"loaded {num_documents} documents")
```

```python
folder_path = '/content/drive/MyDrive/Nvidia_Data/llm'
```

```python
import os
```

```python
for i, doc in enumerate(docs):
    file_path = os.path.join(folder_path, f'document_{i + 1}.html')
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(doc.page_content)

print(f"Saved {num_documents} documents in {folder_path}")
```
